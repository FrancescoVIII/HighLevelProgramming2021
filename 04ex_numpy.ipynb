1\. Find the total mean, and the mean for each row and column of the following matrix:

2\. Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
vect_out = []

print("Using outer in numpy: ")
out_np = np.outer(u,v)
print(out_np)
print("\n")

print("Using a nested for: ")
for i in u:
    for j in v:
        vect_out += [i*j,] 
        
vect_out = np.array(vect_out).reshape((4,4))        
print(vect_out)
print("\n")

print("Using a list comprehension: ")
list_out = [(i*j) for i in u for j in v]
list_out = np.array(list_out).reshape((4,4))  
print(list_out)
print("\n")

print("Using a numpy broadcasting operations: ")
a = np.tile(u, (4, 1)).T
print(a*v)

````

5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anti-diagonal matrix (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagonal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
u = np.arange(1,11)
a = np.tile(u, (10, 1)).T
mult_tb = a*u
print("The multiplication table: ")
print(mult_tb)
print("\n")

print("Extract the anti diagonal: ")
print(np.fliplr(mult_tb).diagonal())
print("\n")

print("Extract the diagonal offset by 1 upwards: ")
print(np.diag(mult_tb,1))

````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import time

def prime_N(N):
    
    x = np.arange(0,N)
    condition = []
    st_t = time.time() #Strat time
    for i in range(N):  #Creating the mask
   
        bl = True
        for j in range(2,int(x[i]/2)+1):
            if(x[i] % j == 0):
                bl = False
        condition += [bl,]
     
    f_t = time.time() - st_t
    
    return x[condition], f_t

def prime_eratosthenes(N):
    st_t = time.time() #Strat time
    prime_list = []
    prime = np.arange(0,N)
    condition = []
    condition += [True,] #I need to add the fist position (0), to have the same size of the array
    for i in range(2, N+1):
        bl = False
        if i not in prime_list:
            bl = True
            for j in range(i*i, N+1, i):
                prime_list.append(j)
        condition += [bl,]
        
    f_t = time.time() - st_t
    
    return prime[condition], f_t


N = [99, 1000, 10000, 50000]

print("| NORMAL ALGORITHM |")
for i in N:
    x, f_t = prime_N(i)
    
    print("The prime number for N=" + str(i) + " are the following: ")
    print(x)
    if (i==50000):
        f_t = f_t/60
        print("The time needed to perform them was: " + "%0.3f"%f_t + " minutes")
        print("\n")
    else:
        print("The time needed to perform them was: " + "%0.3f"%f_t + " seconds")
        print("\n")
    
print("| ERATOSTHENES ALGORITHM |")
for i in N:
    x, f_t = prime_eratosthenes(i)
    
    print("The prime number for N=" + str(i) + " are the following: ")
    print(x)
    print("Time needed to perform them was: " + "%0.3f"%f_t + " seconds")
    print("\n") 
 
    

print("We can notice that the eratosthenes algorithm is more efficent, it has a complexity of o(n), instead \n" 
      + "the normal algorithm has a complexity of o(n^2), so it needs more time to compute the results. \n"
     + "As we can see from the results, for N = 50000 the difference is quite huge")


        
````
